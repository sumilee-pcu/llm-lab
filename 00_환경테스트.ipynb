## 1. 패키지 버전 확인

```python
import google.generativeai as genai
import langchain
import langgraph

print(f'google-generativeai: {genai.__version__}')
print(f'langchain          : {langchain.__version__}')
print(f'langgraph          : {langgraph.__version__}')
```

## 2. .env 로드 & API 키 확인

```python
import os
from dotenv import load_dotenv

load_dotenv()  # .env 파일 로드
api_key = os.getenv('GEMINI_API_KEY')

if api_key:
    print(f'✅ API 키 로드 성공: {api_key[:8]}...{api_key[-4:]}')
else:
    print('❌ API 키를 찾을 수 없습니다. .env 파일을 확인하세요.')
```

## 3. Gemini API 직접 호출 테스트

```python
import google.generativeai as genai

genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.0-flash')

response = model.generate_content('안녕하세요! 한 문장으로 자기소개 해주세요.')
print(response.text)
```

## 4. LangChain + Gemini 연결 테스트

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage

llm = ChatGoogleGenerativeAI(
    model='gemini-2.0-flash',
    google_api_key=api_key,
    temperature=0.3
)

response = llm.invoke([HumanMessage(content='LangChain과 Gemini 연결 성공! 한 줄로 확인 메시지 출력해줘.')])
print(response.content)
```
